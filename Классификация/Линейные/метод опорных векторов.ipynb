В **мультиклассовой** регрессии используют подход __один против всех__  
Суть метода заключается в том, что строится столько бинарных классификаторов, сколько всего есть классов. Каждый классификатор пытается отделить свой класс от остальных.

```python
from sklearn.datasets import make_blobs
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
import numpy as np
import mglearn
```

```python
X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Признак 0")
plt.ylabel("Признак 1")
plt.legend(["Класс 0", "Класс 1", "Класс 2"])
```

```python
lsvm = LinearSVC()
lsvm.fit(X, y)
```

```python
mglearn.plots.plot_2d_classification(lsvm, X, fill=True, alpha=.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(lsvm.coef_, lsvm.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
    plt.legend(['Класс 0', 'Класс 1', 'Класс 2', 'Линия класса 0', 'Линия класса 1', 'Линия класса 2'], loc=(1.01, 0.3))
    plt.xlabel("Признак 0")
    plt.ylabel("Признак 1")
```
